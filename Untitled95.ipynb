TASK-2(Descriptive Stastics and Visualization)

```python
import numpy as np
import pandas as pd
data=pd.read_csv("/content/cleaned_dataset.csv")
data.head()
data.info()
data.describe()
```

```python
df=pd.DataFrame(data)
numerical_df=df.select_dtypes(include=['number'])
mean_values= numerical_df.mean()
median_value=numerical_df.median()
mode_values=numerical_df.mode().iloc[0]
std_dev_values=numerical_df.std()
print("Mean Values:\n", mean_values)
print("Median:\n", median_value)
print("Mode:\n",mode_values)
print("Standard Deviation:\n", std_dev_values)

```

```python
import matplotlib.pyplot as plt
df.hist(figsize=(10,8),bins=20)
plt.suptitle('histogram of Numerical Features')
plt.show()
```

visualization using the Boxplot

```python
df.plot(kind='box',subplots=True,layout=(4,4),figsize=(12,10),sharex=True,sharey=False)
plt.suptitle('Boxplot of the Numerical Features')
plt.show()
```

visualization using the Correlation Matrix and HeatMap

```python
import seaborn as sns
correlation_matrix=numerical_df.corr()
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

pair Plot

```python
sns.pairplot(df)
plt.suptitle('Pair plot of Numerical features')
plt.show
```

```python
df['Rating'].value_counts().plot(kind='bar',figsize=(8,6))
plt.title('Distribution of Rating')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
df['Duration'].value_counts().plot(kind='bar',figsize=(8,6))
plt.title('Distribution of Duration')
plt.xlabel('Duration')
plt.ylabel('Count')
plt.show()
```

Visualization Using the Scatterplot

```python
plt.figure(figsize=(8,6))
plt.scatter(df['Rating'], df['Duration'], alpha=0.5) # Use df['Duration'] to access the column data
plt.title('Rating vs Duration')
plt.xlabel('Rating')
plt.ylabel('Duration')
plt.grid(True)
plt.show()
```
